# Shubert-Piyavskii method

$\newcommand{\absval}[1]{\left\lvert #1 \right\rvert}$

In contrast with other methods, the Shubert Piyavskii method is a *global optimization method* over the domain $[a,b]$, meaning it is guaranteed to converge on the global minimum of a function irrespective of any local minima or whether the function is unimodal. 

From basic analysis, recall the following:

To show directly that $f(x)=3x+1$ is continuous at an arbitrary point  $c \in \mathbf{R}$, we must argue that $\absval{f(x) - f(c)}$ can be made arbitrarily small for values of $x$ near $c$. Now,

\begin{align*}
\absval{f(x) - f(c)} = \absval{(3x + 1) - (3c + 1)} = 3 \absval{x - c}
\end{align*}

So, given, $\epsilon > 0$, we choose $\delta = \epsilon/3$. Then, $\absval{x - c} < \delta$ implies

\begin{align*}
\absval{f(x) - f(c)} = 3 \absval{x - c} < 3 \left(\frac{\epsilon}{3}\right) = \epsilon
\end{align*}

Of particular importance is the fact, that the choice of $\delta$ is the same regardless of which point $c \in \mathbf{R}$, we are considering.

```julia
using Plots

x = range(0, stop=10, length=1000)
f(x) = 3*x + 1
y = x -> f(x)
Plots.plot(x,y)
```

Let's contrast this with what happens when we prove that $g(x) = x^2$ is continuous on $\mathbf{R}$. Given $c \in \mathbf{R}$, we have:

\begin{align*}
\absval{g(x) - g(c)} = \absval{x^2 - c^2} = \absval{x - c} \absval{x + c}
\end{align*}

We need an upper bound on $\absval{x + c}$, which is obtained by insisting that our choice of $\delta$ not exceed $1$. This guarantees that all values of $x$ under consideration will necessarily fall in the interval $(c - 1, c + 1)$. It follows that 

\begin{align*}
\absval{x + c} \le \absval{x} + \absval{c} \le (\absval{c} + 1) + \absval{c} = 2 \absval{c} + 1
\end{align*}

Now, let $\epsilon > 0$. If we choose $\delta = \min \{1, \epsilon/(2 \absval{c} + 1)\}$, then $\absval{x - c} < \delta$ implies:

\begin{align*}
\absval{f(x) - f(c)} = \absval{x - c}\absval{x + c} < \left(\frac{\epsilon}{2 \absval{c} + 1}\right)(2\absval{c} + 1) = \epsilon
\end{align*}

Now, there is nothing deficient about this argument, but it is important to notice that, in the second proof, the algorithm for choosing the response $\delta$ depends on the value of $c$. The statement

\begin{align*}
\delta = \frac{\epsilon}{2 \absval {c} + 1}
\end{align*}

means that larger values of $c$ are going to require smaller and smaller values of $\delta$, a fact that should be evident from a consideration of the graph of $g(x) = x^2$. Given, say $\epsilon = 1$, a $\delta$-response of $\delta=1/3$ is sufficient for $c = 1$, because $2/3 < x <4/3$, certainly implies $0 < x^2 < 2$. However, if $c = 10$, then the steepness of the graph of $g(x)$ means that a much smaller $\delta$ is required $\delta = 1/21$ by our algorithm - to force $99 < x^2 < 101$. 

Recall that to say that $f$ is continuous on $A$ means that $f$ is continuous at each individual point $c \in A$. In other words, given $\epsilon > 0$ and $c \in A$, we can find a $\delta > 0$ *perhaps depending on $c$* such that if $x \in V_\delta(c)$, then $f(x) \in V_\epsilon(f(c))$.

Uniform continuity is strictly stronger property.

**Definition. (Uniform Continuity)**. A function $f:A \to \mathbf{R}$ is uniformly continuous on $A$, if for every $\epsilon > 0$, there exists $\delta > 0$ such that for all pairs of points $x,y \in A$, $\absval{x - y} < \delta$ implies that $\absval{f(x) - f(y) < \epsilon$.

The key distinction between asserting that $f$ is uniformly continuous on $A$ versus simply continuous on $A$ is that, given an $\epsilon > 0$, a single $\delta > 0$ can be chosen that works simultaneously for all points $c$ in $A$. To say that a function is not uniformly continuous on a set $A$ then does not necessarily mean it is not continuous at some point. Rather, it means that there is some $\epsilon_0 > 0$ for which no single $\delta > 0$ is a suitable response for all $c \in A$. 

**Definition. (Lipschitz Functions)**. A function $f:A \to \mathbf{R}$ is called **Lipschitz** if there exists a bound $M > 0$ such that 

\begin{align*}
\absval{\frac{f(x) - f(y)}{x - y}} \le M
\end{align*}

for all $x \neq y$. Geometrically speaking, a function $f$ is Lipschitz if there is a uniform bound on the magnitude of the slope of all lines drawn through any two points on the graph of $f$.

Consider the following. Pick an arbitrary $\epsilon > 0$. Let us explore the expression $\absval{f(x) - f(y)}$.

\begin{align*}
\absval{f(x) - f(y)} \le M \absval{x - y}
\end{align*}

for all pairs of points $x \neq y$. 

Let $\delta = \frac{\epsilon}{M}$. Note that this $\delta$-response does not depend on the value of $x$. Then, whenever $\absval{x - y} < \delta$, we have

\begin{align*}
\absval{f(x) - f(y)} \le M \absval{x - y} < M \cdot \frac{\epsilon}{M} = \epsilon
\end{align*}

So, we have shown that $(\forall \epsilon > 0)$, $(\exists \delta > 0)$, such that $\forall x_1,x_2 \in A$, $\absval{x_1 - x_2} < \delta$ implies $\absval{x_1 - x_2} < \epsilon$. Consequently, $f$ is uniformly continuous. Note however, that the converse is not true. All uniformly continuous functons are not Lipschitz continuous.

```julia
using Plots

x = range(0, stop=10, length=1000)
f(x) = x^2
y = x -> f(x)
Plots.plot(x,y)
```

The Shubert-Piyavskii method requires that the function be *Lipschitz continuous*, meaning that it is continuous and there is an upper bound on the magnitude of its derivative. As mentioned above, a function $f$ is Lipschitz continuous on $[a,b]$ if there exists an $l > 0$ such that:

\begin{align*}
\absval{f(x) - f(y)} \leq l \absval{x - y}
\end{align*}

for all $x,y \in [a,b]$ 

The equation of the straight line passing through the point $(x_0,f(x_0))$ with slope $\pm l$ is given by,

\begin{align*}
y - f(x_0) &= \pm l(x - x_0)\\
y &= \pm l(x - x_0) + f(x_0)
\end{align*}

Therefore, we know that the lines $y = -l(x - x_0) + f(x_0)$ for $x > x_0$ and $y = l(x - x_0) + f(x_0)$ for $x < x_0$ form a lower bound for $f$.

<img src="./images/shubert_piyavskii.png">

That is,

\begin{align*}
f(x) &\geq -l(x - x_0) + f(x_0)\\
f(x) &\geq l(x - x_0) + f(x_0)
\end{align*}

Using these two inequalities, we can construct a piecewise linear sawtooth function

\begin{align*}
\hat{f}(x) \leq f(x), \quad \forall x \in [a,b]
\end{align*}

where $\hat{f}$ is given by,

\begin{align*}
    \hat{f}(x) = 
    \begin{cases} 
    & -l(x - x_0) + f(x_0) & \quad x > x_0 \\
    & l(x - x_0) + f(x_0) & \quad x < x_0 
    \end{cases}
\end{align*}

The Shubert-Piyavskii method iteratively builds a tighter and tighter lower bound on the function. Given a valid Lipschitz constant $l$, the algorithm begins by sampling the midpoint $x^{(1)} = (a + b)/2$. A sawtooth lower bound is constructed using lines of slope $\pm l$ from this point. These lines will always lie below $f$ if $l$ is a valid Lipschitz constant. 

Upper vertices in the sawtooth correspond to sampled points. Lower vertices in correspond to intersections between the Lipschitz lines originating from each sampled point. Further iterations find the minimum point in the sawtooth, evaluate the function at that $x$ value, and then use the result to update the sawtooth. 

<img src="./images/shubert_piyavskii2.png">

A sequence of points $x_1,x_2,x_3,\ldots$ converging to the global maximum is generated by constructing a saw-tooth cover under the function. The algorithm, as I understand it, is presented below.

***Iteration 1.*** In the following, I denote $y_n \equiv f(x_n)$. To start with, a first sampling point $x_1$ is chosen at the mid-point of the interval as $x_1 = (a + b)/2$. At $x_1$ a pair of lines are drawn with slope $L$ given by the equation $y = y_1 + L\absval{x - x_1}$. Note that this pair of lines will not intersect the graph of $f(x)$ owing to $L$ representing the upper bound on the derivative. The intersection of this pair of lines with the end-points of the interval gives us an initial saw-tooth consisting of two points, namely, $\{(t_1,z_1),(t_2,z_2)\}$, where 

\begin{align*}
(t_1,z_1) &= (a,y_1 + \frac{L}{2}(b-a))\\
(t_2,z_2) &= (b,y_1 + \frac{L}{2}(b-a))
\end{align*}

***Iteration k.*** Let the saw-tooth vertices be represented by the list $\{(t_1,z_1),(t_2,z_2),\ldots,(t_n,z_n)\}$. For simplicity, assume that the vertices are arranged in a decreasing order i.e. $z_1 \ge z_2 \ge \ldots \ge z_n$. 

The function now is sampled at the location of the minimum point in the sawtooth. So, we select the $\min \{z_1,z_2,\ldots,z_n\}$ and take the corresponding $t$, and sample the function at this point. In other words,

\begin{align*}
t = \arg \min_{1 \leq i \leq n} \{z_1,z_2,\ldots,z_n\}
\end{align*}

Therefore, this point would be $(t_n,z_n)$. The function is now sampled at the location of the largest peak in the saw-tooth cover, namely at $t_n$, and a pair of lines with slope $L$ are drawn from $y_n = f(t_n)$, whose intersection with the existing saw-tooth cover is given by the vector $\{(t_1,z_1),(t_2,z_2),\ldots,(t_{n-1},z_{n-1}),(t_l,z_l),(t_r,z_r)\}$. These coordinates are sorted in ascending values of the second component. Expressions for the two new points from the geometry of the situation are:

\begin{align*}
z_l = z_r = \frac{y_n + z_n}{2}, \quad t_l = t_n - \frac{1}{2L}(z_n - y_n), \quad t_r = t_n + \frac{1}{2L}(z_n - y_n)
\end{align*}

***Stopping criterion and intervals of uncertainty.*** The algorithm is stopped when the difference in height between the maximum saw-tooth peak and the corresponding function value is less than a specified tolerance, or when 

\begin{align*}
z_n - y_n \le \epsilon
\end{align*}

The regions in which the minimum could lie can be computed using this update information. For every peak, an uncertainty region can be computed according to:

\begin{align*}
    &[t_i - \frac{1}{L}(z_i - y_n), t_i + \frac{1}{L}(z_i - y_n)] \\
    & z_i \ge y_n
\end{align*}

The uncertainty set is a union of all the above intervals.

```julia
struct Pt
    x
    y
end

# Geometrically this function gives the intersection of a pair of lines:
#  L: Straight line with slope -l passing through A(A.x, A.y)
#  L: Straight line with slope l passing through B(B.x, B.y)

# The coordinates of the intersection point are given by
# x = ((A.y - B.y) + l(A.x + B.x))/2l
# y = (A.y + B.y)/2 + l(A.x - B.x)/2

function _get_sp_intersection(A, B, l)
    t = ((A.y - B.y) - l*(A.x - B.x))/(2*l)
    return Pt(A.x + t, A.y - t*l)
end

function shubert_piyavskii(f, a, b, l, )
    x = (a + b)/2
    y = f(x)

    A = Pt(a, f(a))
    M = Pt(x,y)
    B = Pt(b, f(b))

    P = _get_sp_intersection(A, M, l)
    P = _get_sp_intersection(M, B, l)

    pts = [A, P, M, P, B]

     = Inf

    while ( > )
        i = argmin([P.y for P in pts])
        P = Pt(pts[i].x, f(pts[i].x))

         = P.y - pts[i].y

        P = _get_sp_intersection(pts[i-1], P, l)
        P = _get_sp_intersection(P, pts[i+1], l)

        deleteat!(pts, i)
        insert!(pts, i, P)
        insert!(pts, i, P)
        insert!(pts, i, P)
    end
    
    # Every odd point 1,3,5, correspond to upper vertices in the sawtooth.
    i = 2*(argmin([P.y for P in pts[1:2:end]])) - 1
    return pts[i]
end
```

```julia
using Plots

f(x) = x^4 - 4*x^3 - 30*x^2 + 50*x - 100
l = 450
x = range(-5, stop=7, length=1000)
y = x -> f(x)

Plots.plot(x,y)
```

```julia
pts = shubert_piyavskii(f, -5, 7, 450, 0.01)
```
